```python
import numpy as np
import matplotlib.pyplot as plt

# Define the two functions to compare
def pendulo_nao_linear(M, L, G, x):
    return M * L * G * np.sin(x)

def pendulo_linear(M, L, G, x):
    return M * L * G * x

# Setting constants
m = 1
g = 10
l = 1

# Generate x-values
x_values = np.linspace(-(1/2)*np.pi, (1/2)*np.pi, 100)
x_values_specified = [0, np.pi/4, np.pi/2]

# Evaluate both functions at x-values
y1_values = pendulo_nao_linear(m, l, g, x_values)
y2_values = pendulo_linear(m, l, g, x_values)

# Calculate absolute and percentage errors
absolute_errors = np.abs(y2_values - y1_values)
percentage_errors = (absolute_errors / np.abs(y1_values)) * 100

# Plot the graphs
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(x_values, y1_values, label='Function 1: Pendulo nao-linear')
plt.plot(x_values, y2_values, label='Function 2: Pendulo linearizado(x)')
plt.legend()
plt.grid(True)
plt.title('Graphs of Pendulo nao-linear and Pendulo linearizado')

plt.subplot(2, 1, 2)
plt.plot(x_values, absolute_errors, label='Absolute Error')
plt.plot(x_values, percentage_errors, label='Percentage Error')
plt.legend()
plt.grid(True)
plt.title('Errors between Pendulo nao-linear Pendulo linearizado')

plt.tight_layout()
plt.show()

# Output the error values
for x in (0, np.pi/4, np.pi/2):
    erro_absoluto = pendulo_linear(m,l,g,x) - pendulo_nao_linear(m,l,g,x)
    print('Erro absoluto em x = ', x,' =', erro_absoluto)
    erro_relativo  = erro_absoluto / pendulo_nao_linear(m,l,g,x)
    print ('Erro absoluto em x = ', x,' =', erro_relativo * 100, '%')
```
